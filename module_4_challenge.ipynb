## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty

df['line_subtotal'] = round(df['unit_price'] * df['qty'], 2)

df[['unit_price', 'qty', 'line_subtotal']].head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.

import numpy as np

df['total_weight'] = round(df['unit_weight'] * df['qty'], 2)

df['shipping_price'] = np.where(df['total_weight'] > 50, df['total_weight'] * 7, df['total_weight'] * 10)

df[['unit_price', 'unit_weight', 'qty', 'total_weight', 'shipping_price']].head()
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%

df['line_sales_tax'] = round((df['line_subtotal'] + df['shipping_price']) * 0.0925, 2)

df['line_price'] = round(df['line_subtotal'] + df['shipping_price'] + df['line_sales_tax'], 2)

df[['line_subtotal', 'shipping_price', 'line_price', 'line_sales_tax']].head()
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).

df['line_cost'] = round((df['unit_cost'] * df['qty']) + df['shipping_price'], 2)

df[['unit_cost', 'qty', 'shipping_price', 'line_cost']].head()
```

```python
# Create a column for the profit of
# each line using line cost and line price

df['line_profit'] = round(df['line_price'] - df['line_cost'], 2)

df[['line_price', 'line_cost', 'line_profit']].head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.

for id in top5ClientIDs:
    newdf = df.query(f"client_id == {id}")
    print(f"{id}: " + f"${round(newdf['line_price'].sum(), 2)}")
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.

newdf = pd.DataFrame()

for id in top5ClientIDs:
    qdf = df.query(f"client_id == {id}")

    newRow = {"client_id": id, 
              "qty": qdf['qty'].sum(), 
              "shipping_price": "$" + str(round(qdf['shipping_price'].sum() / 100000, 2)) + "M", 
              "line_price": "$" + str(round(qdf['line_price'].sum() / 100000, 2)) + "M", 
              "line_profit": "$" + str(round(qdf['line_profit'].sum() / 100000, 2)) + "M"}
    
    newdf = pd.concat([newdf, pd.DataFrame([newRow])], ignore_index=True)

newdf = newdf.sort_values(by=['line_profit'], ascending=False)

newdf.head()
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.

newdf = newdf.rename(columns={"client_id": "Client ID", "qty": "Units", "shipping_price": "Shipping", "line_price": "Total Revenue", "line_profit": "Total Profit"})

newdf.head()
```
